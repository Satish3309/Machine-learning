```python
#Step-1 (Import Library In Library there are several formulaes are pre loaded to use, 
#so making model creation simple we should import class and functions in our sytem memory)
import pandas as pd 
```

```python
#after import pandas library, we merge our dataset on it
dataset=pd.read_csv("Salary Details.csv")
```

```python
#then next step split input and output from our dataset for learning purpose
independent= dataset[["Yearsof_Experience"]]
```

```python
independent
```

```python
# after that assign our splited input and out put to the recognaised variable so that we can use it anywere we want
dependent= dataset[["Salary"]]
```

```python
dependent
```

```python
# Then split train and test data for learning and test purpose, after spliting assign respective variable to reuse it
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
# In this step we are going to creat model from importing sk learn provider as function , after that creat model with .fit 
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train) #in this step we just put our train qustions and answers to our model
```

```python
# Successfully model created , in this step we are going to review it 
weight=regressor.coef_
weight # this is the slop value which is reson for creating linear line 
```

```python
bais=regressor.intercept_
bais # intercept also known as origion of linear line 
```

```python
#test set- evaluation metrics 
Y_pred=regressor.predict(X_test) # in this step we give splited input for test purpose with out answer.
#so that we can test our model learning is good or not
```

```python
#After test now we should check our model's score for that we use r2. why because we can get answer with out compare two model 
from sklearn.metrics import r2_score #from sklearn library take metrics class and use function (r2 formulae)
r_score=r2_score(Y_test,Y_pred)
```

```python
#to know r2 value just run it (note: if r2 near 1 is good model else poor model)
r_score
```

```python
#To save our best model import pickle 
import pickle 
file_name="finalized_model_linear.sav" # in this step we just create file name only 
```

```python
#After creating file name , now we are goin to write out model to it with write binary command 
pickle.dump(regressor,open(file_name,'wb')) # Now our best model saved successfully in our system memory so that we can use it any time 
```

```python
#Before hiting deployment we should check our trained madel for the input which is not in train and test sets 
#To check our existing good model first we load it from our system memory 
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[13]])
```
